```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def plot_beam(filename):
    intibeam = np.loadtxt(filename)
    intibeam = intibeam[intibeam[:,-2]==0, :]
    intibeam_df = pd.DataFrame(columns=['x','y'])
    intibeam_df['x'] = intibeam[:, 0]
    intibeam_df['y'] = intibeam[:, 2]
    sns.jointplot(x="x", y="y", data=intibeam_df, kind="kde", levels=50, fill=True, cmap='rainbow')
    plt.show()
```

```python
enve_data = pd.read_csv("./beam_envelope", sep=' ')
print(enve_data)
plt.figure(figsize=(13,3))
plt.plot(enve_data.iloc[:,0], enve_data.iloc[:,1])
plt.plot(enve_data.iloc[:,0], enve_data.iloc[:,2])

max_sig = enve_data.iloc[:, [1,2]].max().max()

deviceType_list = ["Dipole", "Solenoid", "Quad"]
for i in range(enve_data.shape[0]):
    if enve_data.iloc[i,4] in deviceType_list:
        begin_position = enve_data.iloc[i-1, 0]
        end_position = enve_data.iloc[i, 0]
        if enve_data.iloc[i,4] == "Dipole":
            plt.fill_between([begin_position, end_position],0, max_sig,facecolor = 'pink', alpha = 0.9)
        elif enve_data.iloc[i,4] == "Solenoid":
            plt.fill_between([begin_position, end_position],0, max_sig,facecolor = 'green', alpha = 0.3)
        elif enve_data.iloc[i,4] == "Quad":
            plt.fill_between([begin_position, end_position],0, max_sig,facecolor = 'blue', alpha = 0.3)   

plt.figure(figsize=(13,6))
plt.plot(enve_data.iloc[:,0], enve_data.iloc[:,3])
```

```python
intibeam = np.loadtxt("initbeam_spacecharge")
plt.figure(figsize=(14,7))
plt.subplot(1,2,1)
plt.scatter(intibeam[:,0], intibeam[:,2], s=1)

endbeam = np.loadtxt("endbeam_spacecharge")
endbeam = endbeam[endbeam[:,-2]==0, :]
plt.subplot(1,2,2)
plt.scatter(endbeam[:,0], endbeam[:,2], s=1)


plot_beam("initbeam_spacecharge")
plot_beam("endbeam_spacecharge")
```

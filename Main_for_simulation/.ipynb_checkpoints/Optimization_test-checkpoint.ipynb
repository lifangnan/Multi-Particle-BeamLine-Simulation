```python
num_particle = 10240

envelope_animation = []
min_loss = 10000
beam_filename_list = []

def modify_parameters_and_simulate(G1, G2, G3):
    simulator.set_magnet_with_name("Q1", G1)
    simulator.set_magnet_with_name("Q2", G2)
    simulator.set_magnet_with_name("Q3", G3)
    simulator.restore_initial_beam()
    envelope = simulator.simulate_and_getEnvelope(False)
    return envelope

def obj(args):
    global min_loss
    global envelope_animation
    global beam_filename_list
    G1, G2, G3 = args
    envelope = modify_parameters_and_simulate(G1, G2, G3)
    
    std_x = envelope[-1, 1]
    std_y = envelope[-1, 2]
    loss_ratio = envelope[-1, 3]/num_particle
    
    ans = std_x + std_y
    
    if min_loss > ans:
        min_loss = ans
        envelope_animation.append(envelope)
        filename = "./beam_file/step_%d"%(len(beam_filename_list))
        simulator.beam_print_to_file(filename)
        beam_filename_list.append(filename)
    else:
        envelope_animation.append(envelope_animation[-1])
        beam_filename_list.append(beam_filename_list[-1])
    return {'loss':ans, 'status': STATUS_OK}
```

```python
envelope_animation = []
beam_filename_list = []
min_loss = 10000
G_trials = Trials()
best_G = fmin(
    fn=obj,
    space=[hp.uniform('G1', 16, 20), hp.uniform('G2', -12.5, -8.5), hp.uniform('G3', 5.4, 9.4)],
    algo=tpe.suggest,
    max_evals=100,
    trials = G_trials)

print(best_G)
plt.plot(G_trials.losses())

vals = [item['misc']['vals'] for item in G_trials.trials]
G1_list = [item['G1'][0] for item in vals]
G2_list = [item['G2'][0] for item in vals]
G3_list = [item['G3'][0] for item in vals]

plt.figure()
plt.plot(G1_list)
plt.plot(G2_list)
plt.plot(G3_list)
plt.show()

envelope = modify_parameters_and_simulate(best_G['G1'], best_G['G2'], best_G['G3'])
simulator.plot_envelope(envelope)

element_types = simulator.get_Beamline_ElementTypes()
element_lengths = simulator.get_Beamline_ElementLengths()
position_start = np.array([])
position_end = np.array([])
for i in range(element_lengths.shape[0]):
    position_start = np.append(position_start, element_lengths[:i].sum())
    position_end = np.append(position_end, position_start[i] + element_lengths[i])
```

# 换评价函数，小搜索范围

```python
def obj_sig(p):
    G1, G2, G3 = p
    envelope = modify_parameters_and_simulate(G1, G2, G3)
    
#     max_x = envelope[-1, 1]
#     max_y = envelope[-1, 2]
    loss_ratio = envelope[-1, 3]/num_particle
    sig_x = simulator.getBeamSigx()
    sig_y = simulator.getBeamSigy()
    
    ans = loss_ratio + (sig_x + sig_y)*10 + abs(sig_x - sig_y)*15

    return ans


GA_model = GA(func=obj_sig, n_dim=3, size_pop=50, lb=[16, -12.5, 5.4], ub=[20, -8.5, 9.4], precision=1e-7, prob_mut = 0.001)
start_time = time.time()
cosumed_time_list = []

for i in range(20):
    best_x, best_y = GA_model.run(10)
    finished_num = (i+1)*10
    cosumed_time = time.time() - start_time
    cosumed_time_list.append(cosumed_time)
    print("耗时%f秒，目前已完成%d代,最小Loss为%f，当前最佳参数为%f, %f, %f"%(cosumed_time, finished_num, best_y, best_x[0], best_x[1], best_x[2]))

# print('parallel mode, time costs: ', (time.time() - start_time))
Y_history = pd.DataFrame(GA_model.all_history_Y)
fig, ax = plt.subplots(2, 1)
ax[0].plot(Y_history.index, Y_history.values, '.', color='red')
Y_history.min(axis=1).cummin().plot(kind='line')
plt.ylabel("loss")
plt.show()

envelope = modify_parameters_and_simulate(best_x[0], best_x[1], best_x[2])
simulator.plot_envelope(envelope)
```

```python
def obj_sig(p):
    G1, G2, G3 = p
    envelope = modify_parameters_and_simulate(G1, G2, G3)
    
#     max_x = envelope[-1, 1]
#     max_y = envelope[-1, 2]
    loss_ratio = envelope[-1, 3]/num_particle
    max_x = envelope[-1, 1]
    max_y = envelope[-1, 2]
    max_x_out = envelope[-2, 1]
    max_y_out = envelope[-2, 2]
    sig_x = simulator.getBeamSigx()
    sig_y = simulator.getBeamSigy()
    
    ans = loss_ratio + (sig_x + sig_y)*10 + abs(max_x - max_y)*4 + abs(max_x_out - max_y_out)*4

    return ans

GA_model = GA(func=obj_sig, n_dim=3, size_pop=100, lb=[0, -25, 0], ub=[25, 0, 25], precision=1e-7, prob_mut = 0.01)
start_time = time.time()
cosumed_time_list = []

for i in range(2):
    best_x, best_y = GA_model.run(10)
    finished_num = (i+1)*10
    cosumed_time = time.time() - start_time
    cosumed_time_list.append(cosumed_time)
    print("耗时%f秒，目前已完成%d代,最小Loss为%f，当前最佳参数为%f, %f, %f"%(cosumed_time, finished_num, best_y, best_x[0], best_x[1], best_x[2]))

# print('parallel mode, time costs: ', (time.time() - start_time))
Y_history = pd.DataFrame(GA_model.all_history_Y)
fig, ax = plt.subplots(2, 1)
ax[0].plot(Y_history.index, Y_history.values, '.', color='red')
Y_history.min(axis=1).cummin().plot(kind='line')
plt.ylabel("loss")
plt.show()

envelope = modify_parameters_and_simulate(best_x[0], best_x[1], best_x[2])
simulator.plot_envelope(envelope)
```

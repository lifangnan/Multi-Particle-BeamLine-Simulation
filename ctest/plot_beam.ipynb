```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_beam(file_name, good_only = False):
    beam_data = np.loadtxt( file_name )
    all_num = beam_data.shape[0]
    if good_only:
        beam_data = beam_data[beam_data[:,-2] != 10, :]
    good_num = beam_data.shape[0]
    print("Good num: %d/%d"%(good_num, all_num))
    
    plt.figure( figsize = (15, 6))
    plt.subplot(1,3,1)
    plt.scatter(beam_data[:,0], beam_data[:,1], s = 1)
    plt.title('x-xp')
    plt.subplot(1,3,2)
    plt.scatter(beam_data[:,2], beam_data[:,3], s = 1)
    plt.title('y-yp')
    plt.subplot(1,3,3)
    plt.scatter(beam_data[:,0], beam_data[:,2], s = 1)
    plt.title('x-y')
    plt.show()
```

```python
plot_beam('initbeam')
```

```python
plot_beam('endbeam', True)
```
